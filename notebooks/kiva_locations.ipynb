```python
from pandas import DataFrame,Series
import pandas
import json
import glob
import os
import collections

def flatten(d, parent_key=''):
    items = []
    for k, v in d.items():
        new_key = parent_key + '_' + k if parent_key else k
        if isinstance(v, collections.MutableMapping):
            items.extend(flatten(v, new_key).items())
        else:
            items.append((new_key, v))
    return dict(items)
```

```python
def get_df():
    i = 0
    for fil in os.listdir('/Users/205250/Documents/dsets/kiva_ds_json/loans/'):
        data = json.loads(open('/Users/205250/Documents/dsets/kiva_ds_json/loans/%s' % fil,'r').read())
        if i == 0:
            df = DataFrame.from_dict([flatten(record) for record in data['loans']])
        else:
            df = pandas.concat([df,DataFrame.from_dict([flatten(record) for record in data['loans']])])
        i+=1
        if i == 10:
            break
    return df 
```

```python
df = get_df()
```

```python
df['location_country'].value_counts()[:40].plot(kind='barh',rot=0)
```
